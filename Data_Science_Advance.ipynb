```python
import pandas as pd
import statsmodels.formula.api as smf

data_retail = pd.read_csv("data_retail.csv", header=0, sep=";")
print(data_retail.head())

model = smf.ols('Average_Transaction_Amount ~ Count_Transaction', data=data_retail)
results = model.fit()
print(results.summary())
```

```python
slope = model.fit().params['Count_Transaction']
intercept = model.fit().params['Intercept']

def Predict_Transaction_Amount(Count_Transaction):
    return (slope * Count_Transaction + intercept)

print(Predict_Transaction_Amount(20))   # prediksi untuk pelanggan dengan 20 transaksi
print(Predict_Transaction_Amount(50))   # prediksi untuk pelanggan dengan 50 transaksi
print(Predict_Transaction_Amount(100))  # prediksi untuk pelanggan dengan 100 transaksi
print(Predict_Transaction_Amount(150))  # prediksi untuk pelanggan dengan 150 transaksi
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data_retail = pd.read_csv("data_retail.csv", header=0, sep=";")

X = data_retail[['Count_Transaction']]
y = data_retail['Average_Transaction_Amount']

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Data aktual')
plt.plot(X, y_pred, color='red', label='Garis Regresi')
plt.xlabel('Count_Transaction')
plt.ylabel('Average_Transaction_Amount')
plt.title('R-Squared Visualization: Linear Regression (Data Retail)')
plt.legend()
plt.show()

r_squared = model.score(X, y)
print("R-Squared:", r_squared)
```
